```python
import numpy as np
from sklearn.decomposition import PCA



print("Loading feature-selected data...")

X_train_selected = np.load("X_train_selected.npy")
X_test_selected = np.load("X_test_selected.npy")

print("\nData Loaded Successfully!")

print("\nTraining Shape :", X_train_selected.shape)
print("Testing Shape  :", X_test_selected.shape)



print("\nApplying PCA...")

pca = PCA(
    n_components=50,
    random_state=42
)

X_train_pca = pca.fit_transform(X_train_selected)
X_test_pca = pca.transform(X_test_selected)

print("\nPCA Completed Successfully!")

print("\nOriginal Features :", X_train_selected.shape[1])
print("PCA Components    :", X_train_pca.shape[1])

print("\nTraining Shape After PCA :", X_train_pca.shape)
print("Testing Shape After PCA  :", X_test_pca.shape)



explained_variance = pca.explained_variance_ratio_.sum()

print("\nTotal Explained Variance:")
print(f"{explained_variance:.4f}")



np.save("X_train_pca.npy", X_train_pca)
np.save("X_test_pca.npy", X_test_pca)

print("\nPCA datasets saved successfully!")

print("Files Generated:")
print("- X_train_pca.npy")
print("- X_test_pca.npy")
```
