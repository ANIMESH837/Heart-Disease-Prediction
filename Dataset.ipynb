```python
import numpy as np
import pandas as pd
import os



print("Generating synthetic high-dimensional dataset...")

np.random.seed(42)

# 500 samples and 10,000 features
X_high_dim = np.random.randn(500, 10000)

# Binary Target Variable
y_high_dim = np.random.choice([0, 1], size=500, p=[0.5, 0.5])

# Feature Names
feature_names = [f"Feature_{i+1}" for i in range(10000)]

# Create DataFrame
df = pd.DataFrame(X_high_dim, columns=feature_names)

# Add Target Column
df["Target"] = y_high_dim



print("\nDataset Created Successfully!")

print("\nDataset Shape:")
print(df.shape)

print("\nFirst 10 Rows and First 20 Features:")

display(df.iloc[:10, list(range(20)) + [-1]])



df.to_pickle("high_dimensional_dataset.pkl")

print("\nDataset saved successfully!")
print("File: high_dimensional_dataset.pkl")

# Verify
print("\nCurrent Working Directory:")
print(os.getcwd())
```
