```python
import numpy as np

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

print("Loading Preprocessed Data...")

X_train_scaled = np.load("X_train_scaled.npy")
X_test_scaled = np.load("X_test_scaled.npy")

y_train = np.load("y_train.npy")
y_test = np.load("y_test.npy")

print("\nData Loaded Successfully!")

print("\nTraining Shape :", X_train_scaled.shape)
print("Testing Shape  :", X_test_scaled.shape)

print("\nStarting Feature Selection...")

selector = SelectKBest(
    score_func=f_classif,
    k=8
)

X_train_selected = selector.fit_transform(
    X_train_scaled,
    y_train
)

X_test_selected = selector.transform(
    X_test_scaled
)

print("\nFeature Selection Completed!")

print("\nOriginal Features :", X_train_scaled.shape[1])
print("Selected Features :", X_train_selected.shape[1])

print("\nTraining Shape After Selection :",
      X_train_selected.shape)

print("Testing Shape After Selection  :",
      X_test_selected.shape)

# Save Selected Features

np.save(
    "X_train_selected.npy",
    X_train_selected
)

np.save(
    "X_test_selected.npy",
    X_test_selected
)

print("\nSelected Feature Datasets Saved Successfully!")

print("\nFiles Generated:")
print("- X_train_selected.npy")
print("- X_test_selected.npy")
```
